```python
# 读取CSV文件
csv_file = "./Data/SVI_withinbuffer.csv"
df = pd.read_csv(csv_file)

# 显示前几行数据
df
```

```python

# 设置源文件夹和目标文件夹
source_folder = "./images1"
target_folder = "./image_SVI_buffer"
# 如果目标文件夹不存在，则创建它
if not os.path.exists(target_folder):
    os.makedirs(target_folder)
```

```python
# 遍历DataFrame中的SVI_ID
for svi_id in df["SVI_ID"]:
    # 根据SVI_ID构建图片文件名，此处假设图片格式为.jpg
    image_filename = f"{svi_id}.jpg"
    
    # 构建源文件和目标文件的完整路径
    source_file = os.path.join(source_folder, image_filename)
    target_file = os.path.join(target_folder, image_filename)

    # 检查源文件是否存在，如果存在，则复制到目标文件夹
    if os.path.exists(source_file):
        shutil.copyfile(source_file, target_file)
        print(f"已复制文件: {image_filename}")
    else:
        print(f"文件未找到: {image_filename}")
```
